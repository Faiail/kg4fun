```python
import os
os.chdir("..")
```

```python
import pandas as pd
from src.utils import load_json
```

```python
ROOT = "./reports/metrics"
```

```python
def load_metric_json(pth: str) -> dict:
    data = load_json(pth)
    nested = {
        (k, vi): vv
        for k, v in data.items()
        if isinstance(v, dict)
        for vi, vv in v.items()
    }

    flat = {
        (k, ""): v for k, v in data.items() if not isinstance(v, dict)
    }
    return {**nested, **flat}
    
```

```python
results = dict()
for (root, _, files) in filter(lambda x: x[1] == [], os.walk(ROOT)):
    for file in files:
        metrics = load_metric_json(f"{root}/{file}")
        row_index = *(root.split(ROOT)[1].split("/")[1:]), file.split(".json")[0]
        results.update({row_index: metrics})
results_df = pd.DataFrame(results).T
```

```python
node_type_res = results_df.loc[(slice(None), "node_types", slice(None)), results_df.columns.get_level_values(1).str.contains("fmeasure", na=False) |
    (results_df.columns.get_level_values(1) == "")].dropna(axis=1, how="all")
node_type_res
```

```python
edge_type_res = results_df.loc[(slice(None), "edge_types", slice(None)), results_df.columns.get_level_values(1).str.contains("fmeasure", na=False) |
    (results_df.columns.get_level_values(1) == "")].dropna(axis=1, how="all")
edge_type_res
```
